```python
import pandas as pd
```

```python
df = pd.read_csv("./tr_inmuebles.csv", delimiter="|", dtype={"ENT":object, "MUN":object, 
                                                             "LOC":object, "AGEB":object, "MZA":object,
                                                            "ENT_ADMON":object})
```

```python
df
```

```python
df["admin_1_code"] = df.ENT
df["admin_2_code"] = df.admin_1_code + df.MUN
df["admin_3_code"] = df.admin_2_code + df.LOC
df["admin_4_code"] = df.admin_3_code + df.AGEB
df["admin_5_code"] = df.admin_4_code + df.MZA
```

```python
filter_col = [col for col in df if col.startswith('admin')]
col_id = ["ID_INM", "P4A", "P4B"]
schools_manzanas = df[col_id+filter_col]
```

```python
schools_manzanas[~schools_manzanas.admin_5_code.isnull()]
```

```python
schools_manzanas.to_csv("./schools_manzanas")
```
